### Question - 3 

Now consider a “Term” to mean a 2-gram (two words occurring sequentially) in a document. Modify the programs given to you to compute the TFIDF for each Term (2-gram). Submit the following items:

a.	List of top 15 “Terms” (2-grams) for each document, having the highest TFIDF values. The task of selecting the top 15 terms does not need to be done by the MapReduce program.

b.	Which output – obtained in 1(b) or in 3(a) – better characterizes the documents? Give reasons for your answers.


### Solution

The given programs are modified to compute TFIDF in the Hadoop itself, the output of the phase 3 are of the form as shown below:

```python
# importing libraries

import numpy as np
import pandas as pd
```

```python
with open("PhaseThree_n/phase3_2g_res/part-r-00000") as file1_1:
    data1_1 = file1_1.read()
    data1_1_n= data1_1.split("\n")
    
print(data1_1_n[0])
```

```python
# Creating a dataframe from the file

df_2g = pd.read_csv("PhaseThree_n/phase3_2g_res/part-r-00000", header=None, dtype={1:str})

df_2g.columns=["2gram","Document","n","N","m","TFIDF"]

df_2g.head(10)
```

```python
# The dataframe is grouped, sorted and indexed to show the top 15 TF-IDF words within each document.

df_2g=df_2g.sort_values(["Document","TFIDF"],ascending=[True,False])

df_2g=df_2g.groupby(["Document"], as_index=False).apply(lambda x: x.reset_index(drop = True)).reset_index()

df_2g=df_2g.drop('level_0',axis = 1).set_index('level_1')

df_2g=df_2g.groupby(["Document"]).head(15)

df_2g.index+=1

df_2g
```

```python
# saving the df after sorting per document by TFIDF and having top 15

df_2g.to_csv("DF_2g", sep='\t', encoding='utf-8')
```

Comments on comparing the df_2g (top 15 2-grams per each document) and df_1_n (top 15 1-grams per each document):

It looks like 2-grams characterizes more against 1-grams since the one grams cannot offer much information like: name of the persons involved / the topics of which the book speaks, the era of the book is speaking etc. Moreover, it is much likely to know the style of the author with the 2-grams.
